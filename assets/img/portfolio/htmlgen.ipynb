```python
import numpy as np
import os
import pyperclip
import random
```

```python
files = [[(i,i+'/'+j) for j in os.listdir(i)] for i in list(filter(lambda v: '.' not in v,os.listdir()))]
flatten = lambda t: [item for sublist in t for item in sublist]
files = flatten(files)

random.shuffle(files)
```

```python
text = """
                <div class="grid-item {c} col-sm-12 col-md-6 col-lg-3">
                    <a href="assets/img/portfolio/{a}" title="{b}">
                        <div class="project_box_one">
                            <img src="assets/img/portfolio/{a}" alt="pro1" />
                            <div class="product_info">
                                <div class="product_info_text">
                                    <div class="product_info_text_inner">
                                    </div>
                                </div>
                            </div>
                        </div>
                    </a>
                </div>
"""
```

```python
out = "".join([text.format(a=y,b=x,c=x.lower()) for (x,y) in files])
```

```python

pyperclip.copy(out)
```
